```python
def load_yolo_model(model_path):
    model = YOLO(model_path)
    print('model succussfully loaded')
    return model
```

```python
model = load_yolo_model(model_path)
```

```python
def detect_objects(model, image_path):
    results = model(image_path)
    
    boxes = results[0].boxes
    return boxes
```

```python
folder_path = r'/home/xstats/Documents/Aditya/SupremeCourtComplete/Extracted Images/d2 1.pdf'
```

```python
all_images = os.listdir(r'/home/xstats/Documents/Aditya/SupremeCourtComplete/Extracted Images/d2 1.pdf')
```

```python
detected_class = []
for i in all_images:
    image_path = os.path.join(folder_path,i)
    boxes = detect_objects(model=model,image_path=image_path)
    box_class = boxes.cls
    if 0 in box_class:
        detected_class.append(i)
    
    print(detected_class)
```

```python
def read_text_files(text_files_path):
    texts = []
    for file_name in sorted(os.listdir(text_files_path)):
        if file_name.endswith('.txt'):
            with open(os.path.join(text_files_path, file_name), 'r') as file:
                texts.append(file.read())
    return texts
```
